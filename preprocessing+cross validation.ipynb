```python
import os
os.listdir("../supervised/CSV")
data = pd.read_csv("../supervised/CSV/CarPrice_Assignment.csv")
pd.options.display.max_columns = None
pd.options.display.max_rows = None
#display(data)
data.head()
```

```python
cars = data.copy()
import re
p = re.compile(r'\w+-?\w+')
cars['car_company'] = cars['CarName'].apply(lambda x: re.findall(p, x)[0])

# replacing misspelled car_company names
# volkswagen
cars.loc[(cars['car_company'] == "vw") | (cars['car_company'] == "vokswagen"), 'car_company'] = 'volkswagen'
# porsche
cars.loc[cars['car_company'] == "porcshce", 'car_company'] = 'porsche'
#toyota
cars.loc[cars['car_company'] == "toyouta", 'car_company'] = 'toyota'
#nissan
cars.loc[cars['car_company'] == "Nissan", 'car_company'] = 'nissan'
#mazda
cars.loc[(cars['car_company'] == "maxda")| (cars['car_company'] == "mazda"),'car_company'] = 'mazda'

cars = cars.drop(['CarName','car_ID'],axis=1)

```

```python
cars.columns.tolist()
cars1 =cars[[
 'symboling',
 'car_company',
 'fueltype',
 'aspiration',
 'doornumber',
 'carbody',
 'drivewheel',
 'enginelocation',
 'wheelbase',
 'carlength',
 'carwidth',
 'carheight',
 'curbweight',
 'enginetype',
 'cylindernumber',
 'enginesize',
 'fuelsystem',
 'boreratio',
 'stroke',
 'compressionratio',
 'horsepower',
 'peakrpm',
 'citympg',
 'highwaympg',
 'price',
 ]]

x=cars1.iloc[: , 1:23]
y=cars1.iloc[: , 24:]

cars_categorical = x.select_dtypes(exclude=np.number)
cars_categorical.head()
```

```python
cars_dummies = pd.get_dummies(cars_categorical, drop_first = False)
cars_dummies.head()
```

```python
x=pd.concat([x,cars_dummies], axis=1)
x.head()
```

```python
x=x.drop(x.select_dtypes(include='object'), axis=1)
x.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_columns = x.columns
x = scaler.fit_transform(x)
x = pd.DataFrame(x)
x.columns = x_columns
x.head()
```

```python
# Hyperparameter Tuning Using Grid Search Cross-Validation
# A common use of cross-validation is for tuning hyperparameters of a model. The most common technique is what is called grid search cross-validation.
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from sklearn.feature_selection import RFE
folds = KFold(n_splits=5, random_state=100, shuffle=True)
hyper_params = [{'n_features_to_select': list(range(2, 40))}]
lm = LinearRegression()
lm.fit(x_train,y_train)
rfe=RFE(lm)
model_cv = GridSearchCV(estimator = rfe, param_grid = hyper_params, scoring= 'r2', cv = folds, verbose = 1,return_train_score=True)      

# fit the model
model_cv.fit(x_train, y_train)         
```

```python
# cv_results = pd.DataFrame(model_cv.cv_results_)
# cv_results

print('Best Score: %s' % model_cv.best_score_)
print('Best Hyperparameters: %s' % model_cv.best_params_)
```
