```python
#reference:towardsdatascience.com

# Gradient descent is an iterative optimization algorithm to find the minimum of a function. 
# Here that function is our Loss Function.


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# plt.rcParams['figure.figsize'] = (12.0, 9.0)

import seaborn as sn
sn.set()


import os
print (os.listdir("../supervised"))

data  = pd.read_csv ("../supervised/Simple_linear_regression.csv")

y=data.iloc[ : , 0]
x=data.iloc[: , 1]
plt.scatter(x,y)
plt.show()



```

```python
#building the model based on gradient descent
m=0
c=0

L=0.0001 #learning rate
epoch=10000 # the number of iterations

n=int(len(x))

for i in range (epoch):
 y_pred=m*x+c
 D_m= (-2/n)*sum(x*(y-y_pred))
 D_c=(-2/n)*sum(y-y_pred)
 m=m-L*D_m
 c=c-L*D_c
print (m,c)

y_pred = m*x + c


sn.lmplot(x=('GPA'),y=('SAT'),data=data)  # regression line
```
